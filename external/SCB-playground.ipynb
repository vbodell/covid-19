```python
from pyscbwrapper import SCB
```

```python
help(SCB)
```

```python
scb = SCB('sv')
```

```python
scb.go_down('BE')
scb.info()
```

```python
scb.go_down('BE0101')
scb.info()
```

```python
scb.go_down('BE0101A')
scb.info()
```

```python
scb.go_down('BefolkningR1860')
scb.get_variables()
```

```python
scb.set_query(ålder=scb.get_variables()['ålder'],
              kön=scb.get_variables()['kön'],
              tabellinnehåll=["Folkmängd"], 
              år=["2019"])
scb.get_query()
```

```python
scb.get_url()
```

```python
data = scb.get_data()['data']
data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df['age'] = [k[0] for k in df.key]
df['sex'] = [k[1] for k in df.key]
df['year'] = [int(k[2]) for k in df.key]
df['value'] = [int(k[0]) for k in df['values']]
df
```

```python
scb_sex = scb.get_variables()['kön']
df['sex'] = [scb_sex[int(v)-1] for v in df['sex']]
```

```python
population_2019_by_age = df.loc[:len(df)-3,['age', 'sex', 'value']]
```

```python
# Bin age data
population_2019_by_age.loc[population_2019_by_age.age == '100+','age'] = 100
population_2019_by_age['age'] = population_2019_by_age['age'].astype(int)

lbls = [f'{v}0-{(v+1)*10-1}' for v in range(10)]
lbls[0] = '0-9'
lbls[-1] = '90+'
population_2019_by_age['agegroup'] = pd.cut(population_2019_by_age['age'], 10, precision=3, labels=lbls)
```

```python
population_2019_by_age.to_pickle('pop_2019.pkl')
```
